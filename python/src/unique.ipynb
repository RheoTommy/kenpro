Unique dataset for evaluating the accuracy of the algorithms.

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
from sklearn.datasets import make_blobs
from noise import add_uniform_noise

style.use('ggplot')

X, y = make_blobs(n_samples=1000, centers=4, n_features=2)
X_aug, idx_new = add_uniform_noise(X, p=0.2)
y_aug = np.concatenate([y, -np.ones(len(idx_new), dtype=y.dtype)])

np.savetxt("../out/blobs.csv", X_aug, delimiter=",", comments='')
plt.scatter(X_aug[:, 0], X_aug[:, 1], c=y_aug, s=20, cmap='viridis')
plt.show()
plt.clf()
```

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.1)
X_aug, idx_new = add_uniform_noise(X, p=0.2)
y_aug = np.concatenate([y, -np.ones(len(idx_new), dtype=y.dtype)])

np.savetxt("../out/moons.csv", X_aug, delimiter=",", comments='')
plt.scatter(X_aug[:, 0], X_aug[:, 1], c=y_aug, s=20, cmap='viridis')
plt.show()
plt.clf()
```

```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples = 1000, noise = 0.05, factor = 0.5)
X_aug, idx_new = add_uniform_noise(X, p=0.2)
y_aug = np.concatenate([y, -np.ones(len(idx_new), dtype=y.dtype)])

np.savetxt("../out/circles.csv", X_aug, delimiter=",", comments='')
plt.scatter(X_aug[:, 0], X_aug[:, 1], c=y_aug, s=20, cmap='viridis')
plt.show()
plt.clf()
```
